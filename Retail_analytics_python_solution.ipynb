```python
### Perform outlier treatment by capping the outliers to the upper and lower limits
```

```python
## FOR INCOME
```

```python
Q1 = data['Income'].quantile(0.25)
Q3 = data['Income'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```python
## Capping
```

```python
data['Income_capped'] = np.where(
    data['Income'] > upper_limit, upper_limit,
    np.where(data['Income'] < lower_limit, lower_limit, data['Income']))
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.histplot(data['Income'], bins=50, color='darkviolet')
plt.title("Before Outlier Treatment")
plt.xticks(fontsize=8)

plt.subplot(1,2,2)
sns.histplot(data['Income_capped'], bins=50, color='lightseagreen')
plt.title("After Outlier Treatment")
plt.xticks(fontsize=9)

plt.show()
```

```python
## FOR AGE
```

```python
Q1 = data['Age'].quantile(0.25)
Q3 = data['Age'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```python
## Capping

data['Age_capped'] = np.where(
    data['Age'] > upper_limit, upper_limit,
    np.where(data['Age'] < lower_limit, lower_limit, data['Age']))
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.histplot(data['Age'], bins=50, color='chocolate')
plt.title("Before Outlier Treatment")
plt.xticks(fontsize=9)

plt.subplot(1,2,2)
sns.histplot(data['Age_capped'], bins=50, color='teal')
plt.title("After Outlier Treatment")
plt.xticks(fontsize=9)

plt.show()
```

```python
## 8.Hypothesis Testing: Define and test hypotheses related to customer behavior using
## statistical tests (stats.ttest_ind()).
```

```python
## Defining a hypothesis: Do married and single customers have different average spending?
## Null Hypothesis (H₀): Mean spending (married) = Mean spending (single)
## Alternative Hypothesis (H₁): Mean spending (married) ≠ Mean spending (single)
```

```python
married = data[data['Marital_Status'] == 'Married']['Total_Spending']
single = data[data['Marital_Status'] == 'Single']['Total_Spending']
```

```python
t_stat, p_val = stats.ttest_ind(married, single, nan_policy='omit')

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
## Result: p-value < 0.05 → reject H₀ → significant difference, means there is significant difference between 
## single and married spending.
```
